```python
print("Tipos de variables en df_t1_05:")
print(df_t1_05.dtypes)
print("\nTipos de variables en df_t1_25:")
print(df_t1_25.dtypes)
```

```python
# Hago que la clase de las variables sean las mismas entre las bases 
for col in df_t1_25.columns:
    if col in df_t1_05.columns:
        df_t1_05[col] = df_t1_05[col].astype(df_t1_25[col].dtype)
```

```python
# Seleccionar variables numéricas
numeric_cols = df_t1_05.select_dtypes(include=['int64', 'float64']).columns

# Estadísticos básicos
print("=== Estadísticos básicos ===")
print(df_t1_05[numeric_cols].describe())

# Detección de valores atípicos usando IQR
print("\n=== Valores atípicos (por IQR) ===")
for col in numeric_cols:
    Q1 = df_t1_05[col].quantile(0.25)
    Q3 = df_t1_05[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df_t1_05[(df_t1_05[col] < Q1 - 1.5 * IQR) | (df_t1_05[col] > Q3 + 1.5 * IQR)][col]
    print(f"{col}: {len(outliers)} valores atípicos")
```

```python
# Elimino observaciones con edades e ingresos negativos
df_t1_05 = df_t1_05[(df_t1_05['ch06'] >= 0) & (df_t1_05['ipcf'] >= 0)]
df_t1_25 = df_t1_25[(df_t1_25['ch06'] >= 0) & (df_t1_25['ipcf'] >= 0)]
```
